```python
import pandas as pd
```

```python
df = pd.read_csv('staff_records.csv')   
```

```python
df_2 = df[~(df['salary'] < 0) & ~(df["job_title"] == None)]
```

```python
average_salary = df_2['salary'].mean()
```

```python
print(average_salary)
```

```python
max_age = df_2['age'].max()
```

```python
print(max_age)
```

```python
pivot_table = pd.pivot_table(df_2, values='salary', index='job_title', columns='age', aggfunc='mean')
```

```python
print(pivot_table)
```

```python
import matplotlib.pyplot as plt
plt.plot(df_2['age'], df_2['salary'], 'o-')

plt.xlabel('age')
plt.ylabel('salary')
plt.title('Залежність зарплати від віку')

plt.show()
```
